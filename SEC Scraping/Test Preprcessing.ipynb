```python
with open("TGT_XML\\2010-02-23_index_969.txt") as file:
    text = file.read()
```

```python
import re
remove = re.sub(r'<(.|\n)*?>' , '', text)
remove = re.sub(r'GRAPHIC(.|\n)*?end' , '', remove)
remove = re.sub(r'&.*?;', '', remove)
remove =  re.sub(r'[^a-zA-Z ]', ' ', remove.lower())
remove =  re.sub(r'\s+', ' ', remove)
```
